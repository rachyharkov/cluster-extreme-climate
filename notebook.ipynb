```python
import pandas as pd
import numpy as np

column_name = ['nama_stasiun', 'rr', 'tgl_pengamatan_ch', 'tx','tgl_pengamatan_tx','tn','tgl_pengamatan_tn','tAvg','tgl_pengamatan_tAvg','rh','tgl_pengamatan_rh','wx','tgl_pengamatan_wx']
dataset = pd.read_csv('extreme-climate-dataset.csv', sep=';', header=None, skiprows=1, names=column_name)

dataset
```

```python
dataset.drop(['tgl_pengamatan_ch', 'tgl_pengamatan_tx','tgl_pengamatan_tn','tgl_pengamatan_tAvg','tgl_pengamatan_rh','tgl_pengamatan_wx'], axis=1, inplace=True)
dataset
```

- 8888: data tidak terukur
- 9999: Tidak Ada Data (tidak dilakukan pengukuran)

```python
# delete row with 8888 value
dataset.drop(dataset[dataset['tx'] == 8888].index, inplace=True)
dataset.drop(dataset[dataset['tx'] == 9999].index, inplace=True)
dataset
```

```python
# preprocessing
dataset = dataset.dropna()
dataset
```

```python
# find coordinate

import json
import ssl
from urllib import request

arr_link_request = [
    'https://gis.bmkg.go.id/arcgis/rest/services/Distribusi_Stasiun/MapServer/0/query?text=Meteorologi&f=json',
    'https://gis.bmkg.go.id/arcgis/rest/services/Distribusi_Stasiun/MapServer/1/query?text=Klimatologi&f=json',
    'https://gis.bmkg.go.id/arcgis/rest/services/Distribusi_Stasiun/MapServer/2/query?text=Geofisika&f=json',
    'https://gis.bmkg.go.id/arcgis/rest/services/Distribusi_Stasiun/MapServer/3/query?text=Stasiun&f=json',
    'https://gis.bmkg.go.id/arcgis/rest/services/Distribusi_Stasiun/MapServer/4/query?text=Balai+Besar&f=json',
]

list_of_allstasiun = []

# ignore ssl certificate
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

for link in arr_link_request:
    with request.urlopen(link, context=ctx) as url:
        data = json.loads(url.read().decode())
        
        for stasiun in data['features']:
            list_of_allstasiun.append(stasiun['attributes'])
            list_of_allstasiun[-1]['x'] = stasiun['geometry']['x']
            list_of_allstasiun[-1]['y'] = stasiun['geometry']['y']

df_allstasiun = pd.DataFrame(list_of_allstasiun)
column_stasiun = ['nama_stasiun', 'longitude', 'latitude']
# change df_allstasiun column name to column_stasiun
df_allstasiun.columns = column_stasiun
df_allstasiun
```

```python
for index, row in dataset.iterrows():
    for index2, row2 in df_allstasiun.iterrows():
        if row['nama_stasiun'] == row2['nama_stasiun']:
            dataset.at[index, 'longitude'] = row2['longitude']
            dataset.at[index, 'latitude'] = row2['latitude']

dataset
```

```python
# drop row with nan value
dataset = dataset.dropna()
dataset
```

```python
# world map
import folium
from folium.plugins import HeatMap

# create map
m = folium.Map(location=[-2.5, 118], zoom_start=5)

lat = dataset['latitude'].tolist()
lon = dataset['longitude'].tolist()

# add circle marker
for i in range(0, len(lat)):
    folium.CircleMarker([lat[i], lon[i]], radius=2, color='red', fill_color='red').add_to(m)
# show map
m
```

```python
# k-means clustering

from sklearn.cluster import KMeans

# select feature
X = dataset[['rr', 'tAvg', 'rh', 'wx']].values
```

```python
# MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt

distortions = []
K = range(1, 10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X_scaled)
    kmeanModel.fit(X_scaled)
    distortions.append(sum(np.min(cdist(X_scaled, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X_scaled.shape[0])

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python

kmeans = KMeans(n_clusters=3).fit(X_scaled)
y_kmeans = kmeans.fit_predict(X_scaled)

# create map
peta_persebaran = folium.Map(location=[-2.5, 118], zoom_start=5)

color = ['red','yellow','green']

# add cluster
for i in range(0, len(y_kmeans)):
    clusterny = y_kmeans[i]
    folium.CircleMarker(
        location=[lat[i], lon[i]],
        radius=13,
        color=color[clusterny],
        fill=True,
        fill_color=color[clusterny],
        fill_opacity=0.7,
        popup='Cluster ' + str(y_kmeans[i]) + ' ' + str(dataset['nama_stasiun'].iloc[i])
    ).add_to(peta_persebaran)

# show map
peta_persebaran
```

```python
# entropy

import math

def entropy(labels):
    n_labels = len(labels)
    if n_labels <= 1:
        return 0
    counts = np.bincount(labels)
    probs = counts / n_labels
    n_classes = np.count_nonzero(probs)
    if n_classes <= 1:
        return 0
    ent = 0.
    for i in probs:
        ent -= i * math.log(i, n_classes)
    return ent

# calculate entropy
entropy(y_kmeans)
```

```python
# count cluster of each cluster
np.bincount(y_kmeans)
```
