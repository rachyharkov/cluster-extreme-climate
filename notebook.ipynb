```python
import pandas as pd
import numpy as np

column_name = ['nama_stasiun', 'rr', 'tgl_pengamatan_ch', 'tx','tgl_pengamatan_tx','tn','tgl_pengamatan_tn','tAvg','tgl_pengamatan_tAvg','rh','tgl_pengamatan_rh','wx','tgl_pengamatan_wx']
dataset = pd.read_csv('extreme-climate-dataset.csv', sep=';', header=None, skiprows=1, names=column_name)

dataset
```

## Preprocessing

```python
dataset.dtypes
```

```python
dataset.drop(['tgl_pengamatan_ch', 'tgl_pengamatan_tx','tgl_pengamatan_tn','tgl_pengamatan_tAvg','tgl_pengamatan_rh','tgl_pengamatan_wx'], axis=1, inplace=True)
dataset
```

```python
dataset.describe()
```

- 8888: data tidak terukur
- 9999: Tidak Ada Data (tidak dilakukan pengukuran)

```python
# delete row with 8888 value
dataset.drop(dataset[dataset['tx'] == 8888].index, inplace=True)
dataset.drop(dataset[dataset['tx'] == 9999].index, inplace=True)
dataset
```

```python
# preprocessing
dataset = dataset.dropna()
dataset
```

```python
dataset.describe()
```

```python
classify = ['TINGGI', 'SEDANG', 'RENDAH']
dataset_classified = dataset.copy()
dataset_classified['rr_level'] = pd.cut(dataset['rr'], bins=3, labels=classify)
dataset_classified['tx_level'] = pd.cut(dataset['tx'], bins=3, labels=classify)
dataset_classified['tn_level'] = pd.cut(dataset['tn'], bins=3, labels=classify)
dataset_classified['tAvg_level'] = pd.cut(dataset['tAvg'], bins=3, labels=classify)
dataset_classified['rh_level'] = pd.cut(dataset['rh'], bins=3, labels=classify)
dataset_classified['wx_level'] = pd.cut(dataset['wx'], bins=3, labels=classify)
dataset_classified.drop(['rr', 'tx', 'tn', 'tAvg', 'rh', 'wx'], axis=1, inplace=True)
dataset_classified
```

```python
# find coordinate

import json
from urllib import request

arr_link_request = [
    'https://gis.bmkg.go.id/arcgis/rest/services/Distribusi_Stasiun/MapServer/0/query?text=Meteorologi&f=json',
    'https://gis.bmkg.go.id/arcgis/rest/services/Distribusi_Stasiun/MapServer/1/query?text=Klimatologi&f=json',
    'https://gis.bmkg.go.id/arcgis/rest/services/Distribusi_Stasiun/MapServer/2/query?text=Geofisika&f=json',
    'https://gis.bmkg.go.id/arcgis/rest/services/Distribusi_Stasiun/MapServer/3/query?text=Stasiun&f=json',
    'https://gis.bmkg.go.id/arcgis/rest/services/Distribusi_Stasiun/MapServer/4/query?text=Balai+Besar&f=json',
]

list_of_allstasiun = []

for link in arr_link_request:
    with request.urlopen(link) as url:
        data = json.loads(url.read().decode())
        for stasiun in data['features']:
            list_of_allstasiun.append(stasiun['attributes'])
            list_of_allstasiun[-1]['x'] = stasiun['geometry']['x']
            list_of_allstasiun[-1]['y'] = stasiun['geometry']['y']

df_allstasiun = pd.DataFrame(list_of_allstasiun)
column_stasiun = ['nama_stasiun', 'longitude', 'latitude']
# change df_allstasiun column name to column_stasiun
df_allstasiun.columns = column_stasiun
df_allstasiun
```
